```python
y_rs = []
for rs in datars:
    i = rs["spearman"]
    for j in range(len(i)):
        i[j] = i[j][28:42]
        i[j] = float (i[j])
    i = i.tolist()
    y_rs.append(i)
    
y_rs_mean=[0 for x in range(0,len(y_rs[0]))]
for y in y_rs:
    for j in range(0,len(y)):
        y_rs_mean[j]+=y[j]
for i in range(0,len(y_rs_mean)):
    y_rs_mean[i]/=len(y_rs)

y_rs_sd = [0 for x in range(0,len(y_rs[0]))]
for y in y_rs:
    for j in range(0,len(y)):
        y_rs_sd[j]+= math.pow((y[j]- y_rs_mean[j]),2)
for i in range(0,len(y_rs_mean)):
    y_rs_sd[i] = math.pow((y_rs_sd[i]/len(y_rs)),0.5)
    
y_rs_less = [0 for x in range(0,len(y_rs[0]))]
y_rs_more = [0 for x in range(0,len(y_rs[0]))]
for i in range(len(y_rs_mean)):
    y_rs_less[i] = y_rs_mean[i] - y_rs_sd[i]
    y_rs_more[i] = y_rs_mean[i] + y_rs_sd[i]
```
